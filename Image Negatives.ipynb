# Image Negatives

The negative of an image with intensity levels in the range is obtained by using the negative transformation 
given by the expression

s = L - 1 - r

This type of processing is particularly
suited for enhancing white or gray detail embedded in dark regions of an
image, especially when the black areas are dominant in size.

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
Originalimage=cv.imread("imagenegative.tif")
```

```python
print("Accessing the image shape so that loops can be run: ",Originalimage.shape)
#print(image[0][0])
#Displaying the image
plt.imshow(Originalimage)
plt.title("Original digital mammogram")
plt.axis("OFF")
```

```python
L=256

for i in range(571):
    for j in range(482):
        r=Originalimage[i][j]
        s=L-1-r
        #print(r)
        #print(s)    Checking whether the function works correctly or not
        Originalimage[i][j]=s
        
        
ImageNegative=Originalimage      
plt.imshow(ImageNegative)
plt.title("Negative image obtained using the negative transformation")
plt.axis("OFF")
```
