```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
```

```python
einsteinimage=cv.imread('einstein.tif')
noisyImage=cv.imread('noisy.tif')
```

```python
noisyImage=cv.resize(noisyImage,(500,500))
einsteinimage=cv.resize(einsteinimage,(500,500))
```

```python
newImage=(einsteinimage+noisyImage)
```

```python
plt.subplot(131)
plt.imshow(einsteinimage)

plt.subplot(132)
plt.imshow(noisyImage)

plt.subplot(133)
plt.imshow(newImage)

```

```python
newImage=(einsteinimage-noisyImage)
plt.subplot(131)
plt.imshow(einsteinimage)

plt.subplot(132)
plt.imshow(noisyImage)

plt.subplot(133)
plt.imshow(newImage)
```

```python
newImage=(einsteinimage*noisyImage)
plt.subplot(131)
plt.imshow(einsteinimage)

plt.subplot(132)
plt.imshow(noisyImage)

plt.subplot(133)
plt.imshow(newImage)
```

```python
newImage=(einsteinimage/noisyImage)
plt.subplot(131)
plt.imshow(einsteinimage)

plt.subplot(132)
plt.imshow(noisyImage)

plt.subplot(133)
plt.imshow(newImage)
```

```python

boneImage=cv.imread('bones.tif')
noisyImage=cv.imread('noisy.tif')


boneImage=cv.cvtColor(boneImage, cv.COLOR_BGR2RGB)
noisyImage=cv.cvtColor(noisyImage, cv.COLOR_BGR2RGB)
boneImage=cv.resize(boneImage,(500,500))
#noisyImage = np.reciprocal(noisyImage)
noisyImage=cv.resize(noisyImage,(500,500))
newImage=((boneImage*.7)-(noisyImage*.3))
plt.subplot(131)
plt.imshow(boneImage)

plt.subplot(132)
plt.imshow(noisyImage)

plt.subplot(133)
plt.imshow(newImage)



```
