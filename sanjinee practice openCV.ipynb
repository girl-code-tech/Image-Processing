# Converting Image to GrayScale
A color image consists of 3 channel depth. But when we are using grayscaling, ultimately the depth of the image is reduced to 1 channel. Hence overall model complexity is reduced. For applications like edge detection, photo sketch, cartooning image we generally use grayscale converted images.

# Rotate Image
The syntax of getRotationMatrix2D() is: cv2.getRotationMatrix2D(center, angle, scale)
Here the center is the center point of rotation, the angle is the angle in degrees and scale is the scale property which makes the image fit on the screen.

```python
img=cv.imread("lena.png")
height, width = img.shape[0:2]
img.shape[0:2]
rotationMatrix = cv.getRotationMatrix2D((width/2, height/2), 180, .5)
```

# The next step is to rotate our image with the help of the rotation matrix.

To rotate the image, we have a opencv method named wrapAffine which takes the original image, 
the rotation matrix of the image and the width and height of the image as arguments.

```python
rotatedImage = cv.warpAffine(img, rotationMatrix, (width, height))


plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(122)
plt.imshow(rotatedImage)
plt.title('Rotated Image')
```

# Crop Image
The starting and ending index of the row and column will define the size of the newly created image. 
For example, start from row number 10 till row number 15 will give the height of the image
start from column number 10 until column number 15 will give the width of the image.

```python
startRow = int(height*.15)

startCol = int(width*.15)

endRow = int(height*.85)

endCol = int(width*.85)
plt.subplot(121)
plt.imshow(img)
plt.title('Original Image')
croppedImage = img[startRow:endRow, startCol:endCol]

plt.subplot(122)
plt.imshow(croppedImage)
plt.title('Cropped Image')
cv.waitKey(0)
```

# Smoothing an Image
Smoothing an image intends to improve its quality, though not perfect,
in some scenarios, it can be a substantial change and key for using the image on further processes.

```python

img=cv.imread("oldImage.jpg")
plt.subplot(121)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(122)
blur = cv.blur(img, (5, 5))
plt.imshow(blur)
plt.title("Smooth Image")
```

# Edge detection 
In edge detection, it gives an outline of the image by extracting edges from the image. We are using a canny filter to perform this task.

```python

img = cv.imread('img3.jpeg',0)

edge_det = cv.Canny(img,100,200)
plt.subplot(111)
plt.imshow(edge_det)
```
